```python
# Intern Profile (YOU CAN EDIT THIS)
intern = {
    "name": "Ali",
    "role": "Data Science Intern",
    "skills": ["Python", "Pandas", "Basic Machine Learning", "Data Visualization"],
    "soft_skills": ["Communication", "Teamwork"]
}

# ✅ Predefined technical and behavioral question templates
technical_templates = [
    "What is your experience with {}?",
    "Can you explain how you would use {} in a project?",
    "What challenges have you faced while working with {}?",
    "How do you apply {} in real-world scenarios?",
    "What are the limitations of {}?"
]

behavioral_templates = [
    "Tell me about a time when you demonstrated {}.",
    "How do you handle situations requiring strong {}?",
    "Can you describe a project where your {} made a difference?",
    "What do you do when your {} is tested under pressure?",
    "How has your {} helped you in working with others?"
]

# ✅ Generate questions
def generate_questions(intern):
    tech_qs = []
    for skill in intern["skills"]:
        for template in technical_templates[:1]:  # Only 1 per skill
            tech_qs.append(template.format(skill))

    beh_qs = []
    for skill in intern["soft_skills"]:
        for template in behavioral_templates[:1]:  # Only 1 per skill
            beh_qs.append(template.format(skill))

    return tech_qs[:5], beh_qs[:5]  # Limit to 5 each

# ✅ Generate and print
tech_qs, beh_qs = generate_questions(intern)

print("Technical Interview Questions:\n")
for i, q in enumerate(tech_qs, 1):
    print(f"{i}. {q}")

print("\nBehavioral Interview Questions:\n")
for i, q in enumerate(beh_qs, 1):
    print(f"{i}. {q}")
```
